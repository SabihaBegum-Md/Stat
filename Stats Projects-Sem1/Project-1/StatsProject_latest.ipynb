Question 2: Two plots displaying information about a single categorical variable

Plot-1

```python
#Q2: Plot-1

# Creating a count plot which shows the frequency of each 'property_type' from the dataframe

#The below code calculates the count of each unique property type in the 'property_type' column of the DataFrame 
#selected_columns using the value_counts() method.
property_type_counts = selected_columns['property_type'].value_counts()

#Creates a new figure for the plot with the given sizes
plt.figure(figsize=(10, 6))

#The function 'countplot' from Seaborn library is used to create a count plot which shows the frequency of the 'property type'
sns.countplot(data=df, x='property_type', order=property_type_counts.index, palette='viridis')
#Setting the x label
plt.xlabel('Property Type')
#Setting the y label
plt.ylabel('Count')
#Setting the title of the plot
plt.title('Count plot for frequency of Each Property Type')
# Rotating the x-axis labels for better visibility
plt.xticks(rotation=45)  
plt.tight_layout()
# Add a caption to explain the data and purpose of the line graph
plt.figtext(0.15, 0.01, 'This count plot displays the frequency of property type from the Airbnb listings', ha='left')

#show function below displays the plot
plt.show()
```

Q.4 How many variables do you think you can successfully represent in a visualization? What happens when you exceed this number?

Ans:- The number of variables that can be successfully represented in a visualization depends on the type of visualization, the complexity of the data, and    
     the audience's ability to interpret it. In general:

  i)  Univariate visualization: Several techniques,including bar charts,histograms and pie charts,can be used to successfully represent a single variable.    
  ii) Bivariate visualization: Scatter plots, line charts, and stacked bar charts all frequently depict two variables.
  iii) Multivariate Visualization: It can be difficult to represent three or more variables.Heatmaps, parallel coordinates, and 3D graphs are a few examples 
                                   of techniques that can be employed, however they can rapidly become confusing and challenging to understand.
    

When you exceed the number of variables that can be effectively represented,various issues arise:

  i)  Overplotting: Data points overlap and become indistinguishable, making it difficult to discern patterns.
  
  ii) Visual clutter: As the visualization fills up, it becomes less clear and causes confusion.
  
 iii) Loss of Interpretability: After a certain point, viewers could find it difficult to draw meaningful conclusions from the picture.
  
It is frequently preferable to use data reduction techniques (such as dimensionality reduction) to condense the most crucial information into a more digestible form when dealing with complex data that has a large number of variables.Furthermore, creating multiple, linked visualizations or interactive dashboards can assist in giving the audience a thorough understanding of the data without overwhelming them.
