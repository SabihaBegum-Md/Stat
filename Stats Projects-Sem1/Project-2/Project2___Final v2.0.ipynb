In this step, we identified missing values. Our approach involved assessing the distribution of each column with missing data by employing a distribution code or histogram. Based on the obtained results, we determined whether to fill the missing values with the mean or median. For instance, if the data distribution shows skewness, we opted for median imputation. on the flip side, if the data distribution was symmetrical, we chose mean imputation.


The choice between using the mean or median for filling missing values depends on the distribution of the data. If the distribution is skewed, especially due to the presence of outliers, relying on the mean might be misleading, as it can be influenced by these extreme values. In such cases, the median being less sensitive to outliers will be better to measure the central tendency.

on the flip side, when the distribution is more symmetric and not heavily influenced by outliers, the mean becomes a reliable indicator of the central tendency. It provides a representative value that aligns well with the overall distribution of the data, making it a suitable choice.

Q.1 Describe the binomial distribution, including its parameters and its probability mass function (PMF) ?

A.1 Binomial Distribution: The binomial distribution models the number of successes in a fixed number (n) of independent Bernoulli trials, where each trial has two possible outcomes—success (usually denoted as 1) or failure (denoted as 0).  The key parameters are the probability of success on a single trial (p) and the total number of trials (n).

Parameters:

n: Number of trials.

p: Probability of success on a single trial.

Probability Mass Function (PMF):

The probability mass function of the binomial distribution gives the probability of getting exactly k successes in n trials:

Example:
 
Suppose you are rolling a fair six-sided die (p = 1/6) five times (n=5). 
What is the probability of getting exactly 2 fours?

Using the binomial PMF:

P(X=2) = (5/2))  × (1/6)^2  × (5/6)^3

#### The code below is an example only to see the expected outcome, check the code below and the output of the code for more understanding.

* n = Number of trials
* p = Probability of success
* PMF (pmf_values) = Probability mass function

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Set the parameters
n = 10
p = 0.5

# Generate values for k (number of successes)
k_values = np.arange(0, n+1)

# Calculate the PMF using the binomial distribution formula
pmf_values = binom.pmf(k_values, n, p)

# Plot the PMF
plt.stem(k_values, pmf_values, basefmt='b-', markerfmt='bo', linefmt='b-')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution PMF (n=10, p=0.5)')
plt.show()
```

Q.2 Describe the normal distribution, including its parameters and its probability density function (PDF) ?

A.2 Normal Distribution:  A symmetric bell-shaped curve defines the normal distribution, also known as the Gaussian distribution, which is a continuous probability distribution. The two parameters that fully characterize it are the mean (μ) and the standard deviation (σ).

Parameters: 

μ: Mean, representing the center of the distribution.

σ: Standard deviation, determining the spread or width of the distribution.

Probability Density Function (PDF): 

The probability density function of the normal distribution is given by:

f(x)= 1/(σ√2π) × e^(-〖(x-μ)〗^2/〖2σ〗^2 )

where:

x: A random variable.

e: The base of the natural logarithm.

Example: Consider a normal distribution with μ=50 and σ=10. What is the probability that a randomly selected value is less than 60?

Using the normal distribution:

P (X < 60) = P (Z <  (60-50)/10) = P(Z<1)

Using a standard normal distribution table or calculator, P(Z<1) is approximately 0.8413.
So, the probability is approximately 84.13%.


#### The code below is an example only to see the expected outcome, check the code below and the output of the code for more understanding.

* mu = Mean (average of a probability distribution)
* sigma = Standard Deviation
* pdf =  Probability density function

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 0
sigma = 1

# Generate values for x:
x_values = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)

# Calculate the PDF using the normal distribution formula:
pdf_values = norm.pdf(x_values, mu, sigma)

# Plot the PDF:
plt.plot(x_values, pdf_values, 'b-', label='Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution (mu=0, sigma=1)')
plt.legend()
plt.show()
```

Q.3 Explain the conditions under which the binomial distribution can be approximated by the normal distribution ?

A.3 : - Approximation of Binomial by Normal Distribution:

Conditions:

* Large n: As the number of trials (n) increases, the binomial distribution approaches a normal distribution.
* np and n(1−p) are both sufficiently large: The normal approximation is more accurate when both np and n(1−p) are greater than 5.

Example:Suppose you have a binomial distribution with n=100 trials and p=0.3. Check if it can be approximated by a normal distribution.

np = 100 × 0.3 = 30

n(1 - p) = 100 × 0.7 = 70

Both np and n(1−p) are greater than 5, so the normal approximation is valid.

This allows us to use the normal distribution to estimate probabilities associated with this binomial distribution, simplifying calculations.


#### The code below is an example only to see the expected outcome, check the code below and the output of the code for more understanding.

* n = Number of trials
* p = Probability of success
* PMF = Probability mass function
* k = number of successes

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# Set binomial distribution parameters
n = 30  
p = 0.5 

# Calculate mean and standard deviation of the binomial distribution
mu_binomial = n * p
sigma_binomial = np.sqrt(n * p * (1 - p))

k_values = np.arange(0, n+1)

# Calculate the binomial PMF
pmf_binomial = binom.pmf(k_values, n, p)

# Plot the binomial PMF
plt.bar(k_values, pmf_binomial, label='Binomial Distribution', alpha=0.7)

# Plot the corresponding normal distribution
x_values = np.linspace(0, n, 1000)
pdf_normal = norm.pdf(x_values, mu_binomial, sigma_binomial)
plt.plot(x_values, pdf_normal, label='Normal Distribution Approximation', color='red')

# Add labels and legend
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution and Normal Approximation')
plt.legend()

plt.show()
```

### Observing the convergence of the binomial to the normal distribution

For this part of the question what will do is that we are going to observe the convergence of the binomial to the normal distribution. what we mean by this is each plot will contain 2 plots one is binomial and the other is normal for the same distribution values. that will give us the difference between the two or in other words it will provide us with the convergence.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

n = 10 
p = 0.5
number_of_samples = 1000

num_samples = 1000
binomial_samples = np.random.binomial(n, p, num_samples)

#Clct mean and SD of the binomial distribution
mu_binomial = n * p
sigma_binomial = np.sqrt(n * p * (1 - p))

#normal distribution approximation
normal_samples = np.random.normal(mu_binomial, sigma_binomial, num_samples)


plt.hist(binomial_samples, bins=np.arange(0, n+2)-0.5, density=True, alpha=0.7, label='Binomial Distribution', color='blue', edgecolor='black')
plt.hist(normal_samples, bins=np.arange(0, n+2)-0.5, density=True, alpha=0.7, label='Normal Distribution Approximation', color='orange', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution and Normal Approximation - Histograms')
plt.legend()
plt.show()
```

(n=10)From the plot the blue graph represents the binomial distrubution and the orange one represents the normal distribution approximation.The plot provides a visual comparison between the actual binomial distribution and its normal approximation. In scenarios where the number of trials is sufficiently large, the binomial distribution can be well-approximated by a normal distribution, as suggested by the Central Limit Theorem.

* Central Limit Theorem (CLT): states that the distribution of sample means approximates a normal distribution as the sample size gets larger, regardless of the population's distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm
n = 100
p = 0.5
number_of_samples = 1000
num_samples = 1000

binomial_samples = np.random.binomial(n, p, num_samples)

mu_binomial = n * p
sigma_binomial = np.sqrt(n * p * (1 - p))

#normal distribution approximation
normal_samples = np.random.normal(mu_binomial, sigma_binomial, num_samples)

plt.hist(binomial_samples, bins=np.arange(0, n+2)-0.5, density=True, alpha=0.7, label='Binomial Distribution', color='blue', edgecolor='black')
plt.hist(normal_samples, bins=np.arange(0, n+2)-0.5, density=True, alpha=0.7, label='Normal Distribution Approximation', color='orange', edgecolor='black')

plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution and Normal Approximation - Histograms')
plt.legend()
plt.show()

```

A key idea in statistics, the Central Limit Theorem illustrates how a large number of independently or identically random variables behave as a sum or average. 
It indicates that when the sample size is large enough, the sampling distribution of the sample mean, regardless of the initial distribution shape, moves closer to a normal distribution.
Important things to think about are:

• Sample Size: To achieve a better approximation to the normal distribution, the sample size should be sufficiently large. 

• Independence: The observations have to be independent, or extremely close to it. 

• Distribution: The distribution of the random variables ought to be the same. 

Binomial distribution approximation using a normal distribution and the Central Limit Theorem's significance

• The probability of the distribution of the number of successes is shown by the binomial distribution in a fixed number of independent Bernoulli trials with an equal chance of success (p). 

• The Central Limit Theorem makes it possible for the Binomial distribution to be roughly represented by a normal distribution when the number of trials (n) is high enough and the probability of success is not close to 0 or 1. 

• The variance (µ) and mean (µ) of the binomial distribution will be as follows when the large given sample size (n) and probability of success (p) are not too close to 0 or 1. 

Sample size * success probability = mean | µ = np

σ2 = np (1 - p) | Variance = mean (1-probability of success)

According to the Central Limit Theorem, a large sample size (n) indicates that the sample mean over several samples will resemble the the standard distribution.
