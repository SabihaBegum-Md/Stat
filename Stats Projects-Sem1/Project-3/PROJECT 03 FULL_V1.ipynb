# 2.3 Cleaning the missing values
In this step, we identified missing values. Our approach involved assessing the distribution of each column with missing data by employing a distribution code or histogram. Based on the obtained results, we determined whether to fill the missing values with the mean or median. For instance, if the data distribution shows skewness, we opted for median imputation. on the flip side, if the data distribution was symmetrical, we chose mean imputation.


The choice between using the mean or median for filling missing values depends on the distribution of the data. If the distribution is skewed, especially due to the presence of outliers, relying on the mean might be misleading, as it can be influenced by these extreme values. In such cases, the median being less sensitive to outliers will be better to measure the central tendency.

on the flip side, when the distribution is more symmetric and not heavily influenced by outliers, the mean becomes a reliable indicator of the central tendency. It provides a representative value that aligns well with the overall distribution of the data, making it a suitable choice.

Q.5 Perform a backward feature selection using p-value on these models. 

```python
import statsmodels.api as sm
import pandas as pd

X = df[['accommodates', 'bedrooms']] 
y = df['price']

def backward_feature_selection(X, y, significance_level=0.05):
    features = list(X.columns)
    
    while len(features) > 0:
        X_subset = X[features]
        model = sm.OLS(y, sm.add_constant(X_subset)).fit()
        p_values = model.pvalues[1:]
        max_p_value = p_values.max()
        max_p_value_feature = p_values.idxmax()
        
        # Checking if the highest p-value exceeds the significance level
        if max_p_value > significance_level:
            # Remove the feature with the highest p-value
            features.remove(max_p_value_feature)
        else:
            
            break
    
    return features

selected_features = backward_feature_selection(X, y)
print("Selected Features:", selected_features)

```

```python
X_selected = df[selected_features]
model_selected = sm.OLS(y, sm.add_constant(X_selected)).fit()
print(model_selected.summary())
```

Now, the practical range of your data and the variable modifications you made will determine how meaningful and useful the intercept is. Here are two situations to think about:

#### Interpretation in Practise:

The estimated price at which all variables are at their minimum values is represented by the intercept if the values of the explanatory variables fall within a suitable range that includes zero.

#### Outside the Practical Range:

The intercept may not have a clear, useful interpretation if the zero point is beyond the observed range of your data because of the nature of the modifications (such as squared terms).
In these situations, the intercept is less of a value that correlates to a particular data point and more of a baseline reference point.

### Q5: Perform a backward feature selection using p-value on these models

```python
import statsmodels.api as sm

def backward_feature_elimination(model, X):
    while True:
        p_values = model.pvalues
        max_p_value = p_values.max()
        if max_p_value > 0.05:  
            max_p_value_predictor = p_values.idxmax()
            
            if max_p_value_predictor != 'const':
                X = X.drop(max_p_value_predictor, axis=1)
                X = sm.add_constant(X)  
                model = sm.OLS(model.model.endog, X).fit()
            else:
                # If 'const' is being removed, break the loop
                break
        else:
            # If no predictors have p-values above the threshold, break the loop
            break

    return model


selected_features_model = backward_feature_elimination(nonlinear_model, X_nonlinear)

print(selected_features_model.summary())
```
